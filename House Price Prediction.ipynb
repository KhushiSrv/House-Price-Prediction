```python
data['MSZoning'] = data['MSZoning'].fillna('RL')
```

```python
data['SalePrice'] = data['SalePrice'].fillna(data['SalePrice'].mean())
```

```python
data =data.dropna(axis = 0)
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
data['MSZoning'] = labelencoder.fit_transform(data['MSZoning'])
data['LotConfig'] = labelencoder.fit_transform(data['LotConfig'])
data['BldgType'] = labelencoder.fit_transform(data['BldgType'])
data['Exterior1st'] = labelencoder.fit_transform(data['Exterior1st'])
```

```python
cor_mat = data.corr()

print(data.corrwith(data['SalePrice']).sort_values(ascending= False))
```

```python
data = data.drop(columns = ['BsmtFinSF2', 'Id'])
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
X = data.drop(['SalePrice'], axis = 1)
Y = data['SalePrice']
X_Train, X_Test, Y_Train, Y_Test = train_test_split(X,Y, test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_Train, Y_Train) 
```

```python
Y_pred = model.predict(X_Test)
```

```python
from sklearn.metrics import mean_absolute_percentage_error
mean_absolute_percentage_error(Y_Test, Y_pred)
```
